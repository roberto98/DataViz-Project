```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("stacked.csv")
```

```python
for i in range(len(df)):
    for j in range(len(df.columns)):
        cur_cel = df.iloc[i,j]
        if(type(cur_cel)==str):
            df.iloc[i,j] = cur_cel.replace(u'\xa0', '')

df.to_csv("stacked_fixed.csv")
df
```
